```python
import requests
import json
import pandas as pd
import os
import folium
from folium import plugins
from datetime import datetime
from IPython.display import JSON
```

```python
url = 'https://data.bmkg.go.id/DataMKG/TEWS/gempaterkini.json'
```

```python
response = requests.get(url)
```

```python
gempa_json = response.json()['Infogempa']['gempa']
```

```python
df_gempa = pd.DataFrame.from_records(gempa_json)
```

```python
df_gempa.drop(columns=['Tanggal','Jam','Lintang','Bujur'], inplace=True)
```

```python
df_gempa
```

```python
def konversi_ke_waktu(row):
    return datetime.fromisoformat(row)

df_gempa['DateTime'] = df_gempa['DateTime'].apply(konversi_ke_waktu)
```

```python
df_gempa.info()
```

```python
def utc_to_local(utc):
    return utc.astimezone(tz='Asia/Jakarta')

def deskripsi_waktu(local_date):
    return local_date.strftime('%d %B %Y, %H:%M:%S WIB')

df_gempa['local_date'] = df_gempa['DateTime'].apply(utc_to_local)
df_gempa['deskripsi_waktu'] = df_gempa['local_date'].apply(deskripsi_waktu)
```

```python
def ekstrak_latitude(coordinates):
    return float(coordinates.split(',')[0])

def ekstrak_longitude(coordinates):
    return float(coordinates.split(',')[1])

df_gempa['latitude'] =  df_gempa['Coordinates'].apply(ekstrak_latitude)
df_gempa['longitude'] =  df_gempa['Coordinates'].apply(ekstrak_longitude)
```

```python
def rubah_kedalaman_ke_float(kedalaman):
    return float(kedalaman.replace('km',''))

df_gempa['Kedalaman'] = df_gempa['Kedalaman'].apply(rubah_kedalaman_ke_float)
```

```python
def rubah_magnitude_ke_float(magnitude):
    return float(magnitude)

df_gempa['Magnitude'] = df_gempa['Magnitude'].apply(rubah_magnitude_ke_float)
```

```python
def apakah_berpotensi_tsunami(potensi):
    if potensi == 'Tidak berpotensi tsunami':
        return False
    return True

df_gempa['Potensi'] = df_gempa['Potensi'].apply(apakah_berpotensi_tsunami)
```

```python
df_gempa
```

```python
latitude_indonesia, longitude_indonesia = -2.548926, 118.0148634
```

```python
map_indonesia = folium.Map(location=[latitude_indonesia, longitude_indonesia], zoom_start=5)
```

```python
def potensi_tsunami(potensi):
    return 'Ya' if potensi else 'Tidak'

def buat_popup(row):
    wilayah = '<p>'+row['Wilayah']+'</p><hr>'
    waktu = '<p> Waktu kejadian :'+row['deskripsi_waktu']+'</p><hr>'
    magnitude = '<p> Magnitude :'+str(row['Magnitude'])+' SR</p><hr>'
    kedalaman = '<p> Kedalaman :'+str(row['Kedalaman'])+' km</p><hr>'
    tsunami =  '<p> Berpotensi Tsunami : '+potensi_tsunami(row['Potensi'])+'</p>'
    return '{}{}{}{}{}'.format(wilayah, waktu, magnitude, kedalaman, tsunami)

for index, row in df_gempa.iterrows():
    
    folium.Marker(
      [row['latitude'], row['longitude']], 
      popup = buat_popup(row),
      tooltip = '{} - {} SR'.format(row['deskripsi_waktu'], str(row['Magnitude']))
    ).add_to(map_indonesia)

map_indonesia
```
